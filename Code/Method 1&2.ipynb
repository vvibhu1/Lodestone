```python
import pandas as pd
df = pd.read_csv('Taste_Experiment.csv')
df['sample_item_index'] = df['sample_item_index'].map(str)
```

```python
df.info()
```

```python
## AGGREGATION OF DATA
```

```python
df.drop(['guess_grill_correct'],axis=1,inplace=True)
df_grouped = df.groupby(['grill_type','sample_item_index','item_material','is_frozen']).agg({'thumbs_up_score': ['mean']}).reset_index()
df_grouped.columns = [''.join(col) for col in df_grouped.columns.values]
```

```python
df_grouped
```

```python
# METHOD 1
```

```python
pivot = df_grouped.pivot_table(index=['sample_item_index'], columns=['is_frozen','item_material','grill_type']).fillna(0)
pivot = pivot.set_index(pivot.index.astype(int)).sort_values(['sample_item_index'])
pivot
```

```python
#METHOD 2
```

```python
frozen = df_grouped[df_grouped['is_frozen'] == True].drop(['is_frozen'],axis=1)
non_frozen = df_grouped[df_grouped['is_frozen'] == False].drop(['is_frozen'],axis=1)
```

```python
frozen
```

```python
non_frozen
```

```python
#pivoting both the data frames
```

```python
pivot_frozen = frozen.pivot_table(index=['sample_item_index'], columns=['item_material','grill_type'])
pivot_frozen = pivot_frozen.set_index(pivot_frozen.index.astype(int)).sort_values(['sample_item_index'])
pivot_frozen
```

```python
pivot_non_frozen = non_frozen.pivot_table(index=['sample_item_index'], columns=['item_material','grill_type'])
pivot_non_frozen = pivot_non_frozen.set_index(pivot_non_frozen.index.astype(int)).sort_values(['sample_item_index'])
pivot_non_frozen
```
